## Gradient Descend Alg

```python
def gradient_descend(model, X, spaces, y):    
    # Define the search range for each input
    search_ranges = np.array(spaces)

    # Define the objective function to maximize the output
    def objective(x):
        return -model.predict(x.reshape(1, -1))[0].sum()

    # Define the gradient of the objective function
    def gradient(x):
        h = 0.0001
        grad = np.zeros_like(x)
        for i in range(x.shape[0]):
            xp = np.copy(x)
            xp[i] += h
            grad[i] = (objective(xp) - objective(x)) / h
        return grad

    # Define a function to clip the input to the search range
    def clip(x):
        return np.clip(x, search_ranges[:, 0], search_ranges[:, 1])

    # Perform gradient descent to find the best inputs
    lr = 0.1
    x_init = X
    x_best = clip(x_init)
    y_best = y[0]
    for i in range(1000):
        grad = gradient(x_best)
        x_new = clip(x_best + lr * grad)
        y_new = model.predict(x_new.reshape(1, -1))[0][0]
        if y_new >= y_best:
            return x_new, y_new
        x_best = x_new
        y_best = y_new
```

## Bayesian Alg

```python
from skopt import gp_minimize
from skopt.space import Real, Integer
```

```python

spaces = [Real(0, 82.99989318847656, name="x1"), 
        Real(0, 22607.90234375, name="x2"), 
        Real(0, 755.8787231445312, name="x3"), 
        Real(0, 24.98193359375, name="x4"), 
        Real(0, 84.9219741821289, name="x5"), 
        Real(0, 84.89873504638672, name="x6"), 
        Real(0, 20.802440643310547, name="x7"), 
        Real(0, 0.9607030749320984, name="x8"), 
        Real(0, 37.24669647216797, name="x9"), 
        Real(0, 90, name="x9")]
```

```python
def bayesianAlg(model, X, spaces, y):
    def objective_function(inputs):
        # Reshape the inputs into a 2D array
        inputs = np.array(inputs).reshape(1, -1)

        output = model.predict(inputs)
        
        # Return the negative output, since the optimizer minimizes the objective function
        return -output[0][0]


    # Use Bayesian optimization to find the optimal inputs
    result = gp_minimize(func=objective_function, dimensions=spaces, x0=list(X), y0=y[0], n_calls=50, random_state=0)

    # Print the best inputs and the best output
    return result.x, -result.fun
```

## Genetic Alg

```python
import math as mth
import random
from collections import deque
```

```python
spaces = [(0.0, 82.99989318847656), 
        (0.0, 22607.90234375), 
        (0.0, 755.8787231445312), 
        (0.0, 24.98193359375), 
        (0, 84.9219741821289), 
        (0, 84.89873504638672), 
        (0.0, 20.802440643310547), 
        (0, 0.9607030749320984), 
        (35, 37.24669647216797), 
        (50, 90)]
```

```python
def GeneticAlgorithmWorkHorse(Objfun, real2bin, bin2real, Reproduce, Mutate, rand, nvar, Xreal, Xbin, popsize, lb, ub, ubmlb):
    imax = int(round(popsize/2)); popsize2 = popsize*2; w = -1; AvegVals = deque(); k1 = 0.5; k2 = 0.5; k3 = 0.5; k4 = 0.5;
    pi = mth.pi; Fval = np.zeros(popsize*2); Fits = np.zeros(popsize2); Bfval = mth.inf; sol = np.zeros(nvar); Wfval = -mth.inf;
    
    # Fitness function
    def FitFun(f):
        fitfun = 0.5*(mth.atan(-f)/pi + 1);
        return fitfun;
    
    # Function pointer for sorting
    def takeSecond(elem):
        return elem[1];
    for i in range(popsize):
        # Generate Random Solution and Convert2Binary
        rand(i); real2bin(i);
        # Evaluate Solution
        Fval[i] =  Objfun(i);
        # Compare Solution to Best Solution Ever Found
        if (Fval[i] < Bfval):
            Bfval = Fval[i];
            sol = Xreal[:,i];
        # Compare Solution to Worst Solution Ever Found
        if (Fval[i] > Wfval):
            Wfval = Fval[i]; w = i; 

    # Populate the popsize to 2 popsize
    for i in range(popsize, popsize2):
        Xreal[:,i] = Xreal[:,w]; 
        Xbin[i] = Xbin[w]; 
        Fval[i] = Fval[w];

    # Ranking Initial Population
    Pair = []; Order = [];
    for i in range(popsize2):
        Pair.append((i, Fval[i]));

    Pair.sort(key=takeSecond);
    for i in range(len(Fval)):
        Order.append(Pair[i][0]);
        Fits[i] = FitFun(Pair[i][1]);

    fb = Fits[:popsize].sum() / popsize; fmax = Fits[0]; AvegVals.append(fb);
    iter = 0; maxiter = 800; IsMinimized = 0; Tol = 1e-6;  
    while (iter < maxiter):
        # Selection/Reproduction (Elitism)
        for k in range(imax):
            i = random.randint(0, popsize); j = random.randint(0, popsize);
            fp = max(Fits[i], Fits[j]); pc = k3;
            if(fp >= fb):
                pc = k1 * (fmax - fp) / (fmax - fb);
            if(random.random() < pc):
                Reproduce(i, j, k, Order);

        # Convert to Real, Evaluate and Enqueue
        for i in range(popsize2):
            bin2real(i); Fval[i] = Objfun(i);
            if (Fval[i] < Bfval):
                Bfval = Fval[i]; sol = Xreal[:,i];


        # Sort to keep the next generation at the beginning
        Pair = []; Order = []; 
        for i in range(len(Fval)):
            Pair.append((i, Fval[i]));

        Pair.sort(key=takeSecond);
        for i in range(len(Fval)):
            Order.append(Pair[i][0]);
            Fits[i] = FitFun(Pair[i][1]);

        fb = Fits[:popsize].sum() / popsize; fmax = Fits[0]; AvegVals.append(fb);

        # Genetic Alteration (Mutation)
        for i in range(popsize):
            fi = Fits[i]; pm = k4;
            if(fi >= fb):
                pm = k2 * (fmax - fi) / (fmax - fb);
            if (random.random() < pm):
                Mutate(Order[i]);  

        # Check convergence
        if (iter > nvar):
            AvegmN = AvegVals.popleft();
            IsMinimized = mth.fabs(fb - AvegmN) < Tol;
            if (IsMinimized):
                break;

        iter += 1;
    print("iter = " + str(iter))
    return sol;
```

```python
def GeneticAlgorithm(fun, funInEq, funEq, IntIndx, lb, ub):
    n = len(lb); popsize = 50; penalty = 100; ubmlb = ub - lb; r = [0];
    Xreal = np.zeros((n, popsize*2)); Xbin = []; NotIntIndx = []; 
    
    for i in range(n):
        isin = i in IntIndx;
        if(not isin):
            NotIntIndx.append(i);
    
    for j in range(popsize*2):
        Xbin.append('rand');
    
    for i in range(n):
        l = mth.ceil(mth.log2(ubmlb[i]));
        if (l == 0):
            l = 1;
        if(i in IntIndx):
            r.append(r[i] + l);
        else:
            r.append(r[i] + l + 10);
    
    def Objfun(j):
        Of = fun(Xreal[:, j]); 
        ineq = np.max(0, funInEq(Xreal[:, j])).sum();
        eq = np.abs(funEq(Xreal[:, j])).sum();
        return Of + penalty*(ineq + eq);
    
    def rand(j):
        for i in IntIndx:
            Xreal[i, j] = random.randint(lb[i], ub[i])
        for i in NotIntIndx:
            Xreal[i, j] = lb[i] + random.random()*ubmlb[i];
        
    def real2bin(j):
        Xbin[j] = '';
        for i in NotIntIndx:
            nb = r[i + 1] - r[i]; den = 2**nb;
            xres = mth.floor((Xreal[i, j] - lb[i])*den/ubmlb[i]);
            binary = bin(xres)[2:].zfill(nb);
            Xbin[j] += binary;  
        for i in IntIndx:
            nb = r[i + 1] - r[i];
            xres = Xreal[i, j]  - lb[i];
            binary = bin(xres)[2:].zfill(nb);
            Xbin[j] += binary;  
            
    def bin2real(j):
        for i in NotIntIndx:
            str = Xbin[j][r[i]:r[i + 1]];
            nb = r[i + 1] - r[i]; den = 2**nb - 1;
            xres = int(str, 2);
            Xreal[i, j] = lb[i] + xres*ubmlb[i]/den; 
        for i in IntIndx:
            str = Xbin[j][r[i]:r[i + 1]];
            nb = r[i + 1] - r[i]; den = 2**nb;
            xres = int(str, 2);
            Xreal[i, j] = int(mth.ceil(lb[i] + xres*ubmlb[i]/den)); 
            
    def Reproduce(i, j, k, Order):
        crp = random.randint(1, r[-1]-1);
        Xbin[Order[popsize + k*2]] = Xbin[Order[i]][:crp] + Xbin[Order[j]][crp:];
        Xbin[Order[popsize + k*2 + 1]] = Xbin[Order[j]][:crp] + Xbin[Order[i]][crp:];
    
    def Mutate(i):
        k = random.randint(1, r[-1]-1); kp1 = k+1;
        Xbin[i] = Xbin[i][:k] + str((int(Xbin[i][k]) + 1) % 2) + Xbin[i][kp1:];
        
    ans = GeneticAlgorithmWorkHorse(Objfun, real2bin, bin2real, Reproduce, Mutate, rand, n, Xreal, Xbin, popsize, lb, ub, ubmlb);
    return ans;
```

```python
## Genetic Algorithm prototype

def Maximization_Using_GA(model, X, spaces, y):
    def funInEq(v):
        f = 0;
        return f;

    def funEq(v):
        f = 0;
        return f;

    def objective_function(inputs):
        # Reshape the inputs into a 2D array
        inputs = np.array(inputs).reshape(1, -1)

        output = model.predict(inputs)
        return - output[0][0]
    
    search_ranges = np.array(spaces)
    Lb = search_ranges[:, 0]
    Ub = search_ranges[:, 1]
    optimumvar = GeneticAlgorithm(objective_function, funInEq, funEq, [], Lb, Ub);
    output = objective_function(list(optimumvar)) * -1

    return optimumvar, output



```
