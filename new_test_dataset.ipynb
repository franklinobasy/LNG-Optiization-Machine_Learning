# Generate optimization results

```python
from input_search_algorithm import Optimize
import pickle
import csv

with open('model.pkl', 'rb') as obj:
    model = pickle.load(obj)
```

```python
spaces = [(0.0, 82.99989318847656), 
        (0.0, 22607.90234375), 
        (0.0, 755.8787231445312), 
        (0.0, 24.98193359375), 
        (-0.00830078125, 84.9219741821289), 
        (-0.0166015625, 84.89873504638672), 
        (0.0, 20.802440643310547), 
        (-0.0007421874906867, 0.9607030749320984), 
        (35, 37.24669647216797), 
        (50, 90)]
```

```python
headers = ['MR IGV Position', 'HMR FLOW', 'LMR/HMR RATIO',
       'PERCENTAGE NITROGEN IN MIXED REFERIGERANT',
       'PERCENTAGE METHANE IN MIXED REFERIGERANT',
       'PERCENTAGE ETHANE IN MIXED REFERIGERANT',
       'PERCENTAGE PROPANE IN MIXED REFERIGERANT',
       'PERCENTAGE BUTHANE IN MIXED REFERIGERANT', 'AMBIENT TEMPERATURE',
       'LOW PRESSSURE PROPANE MIXED REFRIGERANT KETTLE LEVEL', 'LNG FLOW', 'LNG RUNDOWN TEMPERATURE',]
```

```python
from tqdm import tqdm
```

```python
with open("new_test_dataset_analysis.csv", 'w', newline='') as f_obj:
    writer = csv.writer(f_obj,)
    writer.writerow(headers)
    for i, row in tqdm(list(data.iterrows())[:50]):
        row = list(row)
        actual_output = np.array(row[:2])
        actual_input = np.array(row[2:])
        actual_prediction = model.predict(actual_input.reshape(1, -1))
        p = Optimize(model.predict, actual_input.copy(), spaces, actual_output.reshape(1, -1))
        result = p.maximize()
        optimized_input = result[0]
        predicted_output = result[1]
        
        writer.writerow(list(actual_input) + list(actual_output) + [actual_prediction])
        writer.writerow(list(optimized_input) + list(predicted_output[0]))

        writer.writerow(list())
```
