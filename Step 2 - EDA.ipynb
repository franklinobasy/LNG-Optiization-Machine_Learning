### Feature Selection by Intuition

By Experience, control variables are variables that can be set by engineers. And from previous knowledge, engineers believe that the following features has impacts on the LNG output: 
- `MR IGV Position`,
- `HMR Flow`, 
- `HMR/MR Ratio`, 
- `Percentage Nitrogen in MR`, 
- `Percentage Methane in MR`, 
- `Percentage Ethane in MR`, 
- `Percentage Propane in MR`,
- `Percentage Butane in MR`, 
- `Ambient Condition`

```python
columns_to_use = ['LNG Flow','LNG Rundown Temperature', 'MR IGV Position','HMR Flow ', 'HMR/MR Ratio', 
                  'Percentage Nitrogen in MR', 'Percentage Methane in MR', 'Percentage Ethane in MR', 
                  'Percentage Propane in MR', 'Percentage Butane in MR', 'Ambient Condition']
```

### Machine learning techniques for feature selection using sklearn

```python
# pearson's correlation feature selection for numeric input and numeric output
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
```

```python
# define feature selection
fs = SelectKBest(score_func=f_regression, k=10)
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
# apply feature selection
X_selected = fs.fit_transform(X, y)
```

```python
# Get columns to keep and create new dataframe with those only
cols = fs.get_support(indices=True)
df2 = df.iloc[:,cols]
df2.head()
```
